## >> Micro-task 3. Matplotlib Library

```python
df.loc[df['College'] == 'Kalyani Government Engineering College']
```

```python
plt.figure(figsize = (16,9))
data = [1,2,1,1,2,1,2,3]

labels = ['Flutter (skill out of 5)', 'Flask (skill out of 5)',
       'Firebase (skill out of 5)', 'GCP (skill out of 5)',
       'AWS (skill out of 5)', 'Predictive Modelling (skill out of 5)',
       'EDA (skill out of 5)', 'NLP (skill out of 5)']

plt.pie(data,labels=labels,autopct = '%0.1f%%', textprops = {'fontsize':13}
       , wedgeprops = {'linewidth':2,'width':1,'edgecolor':'k'},shadow = True)


plt.title('Kalyani Government Engineering College', fontsize = 20)
plt.legend(loc = 3)
plt.show()
```

## >> Micro-task 4. Percentag of NLP in Gitam university, Vizag

```python
df.loc[df['College'] == 'Gitam university, Vizag']
```

```python
plt.figure(figsize = (16,9))
data = [1,1,1,1,3,3,4,2]

labels = ['Flutter (skill out of 5)', 'Flask (skill out of 5)',
       'Firebase (skill out of 5)', 'GCP (skill out of 5)',
       'AWS (skill out of 5)', 'Predictive Modelling (skill out of 5)',
       'EDA (skill out of 5)', 'NLP (skill out of 5)']

plt.pie(data,labels=labels,autopct = '%0.1f%%', textprops = {'fontsize':13}
       , wedgeprops = {'linewidth':2,'width':1,'edgecolor':'k'},explode=(0,0,0,0,0,0,0,0.2),shadow = True)


plt.title('Gitam university, Vizag', fontsize = 20)
plt.legend(loc = 4)
plt.show()
```
