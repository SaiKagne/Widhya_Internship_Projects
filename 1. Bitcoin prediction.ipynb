## >> Micro-task 4. Data Cleaning

```python
plt.figure(figsize = (16,9))
sns.heatmap(df.isnull())
plt.show()
```

```python
df.fillna(df.mean(), inplace = True)
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(df.isnull())
plt.show()
```

## >> Micro-task 5. Model Predication

```python
model = smf.ols(formula='btc_market_price ~ btc_market_cap + btc_n_transactions + btc_miners_revenue + btc_cost_per_transaction + btc_difficulty + btc_hash_rate + btc_cost_per_transaction_percent',data=df).fit()
```

```python
y = model.predict(df[['btc_market_cap','btc_cost_per_transaction_percent', 'btc_n_transactions', 
                      'btc_miners_revenue', 'btc_cost_per_transaction', 'btc_difficulty', 'btc_hash_rate']])

print('MSE:', sm.mean_squared_error(y, df['btc_market_price'].values))
```
